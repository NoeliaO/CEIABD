## Ejercicio 3

Muestra un gráfico de líneas con tres líneas de distintos colores y grosores. Cada una de ellas representará los valores de las funciones trigonométricas seno, coseno y tangente entre los 0 y los 720 grados. Además, muestra un título y las etiquetas (labels):

```python
# Creando un rango de valores desde 0 hasta 720 grados (convertidos a radianes)
x = np.linspace(0, np.radians(720), 400)

# Calculando los valores de las funciones trigonométricas
seno = np.sin(x)
coseno = np.cos(x)
tangente = np.tan(x)

# Limitando los valores de la tangente para mejorar la visualización
tangente = np.clip(tangente, -10, 10)

# Creando el gráfico de líneas para cada función trigonométrica
plt.figure(figsize=(10, 6))
plt.plot(x, seno, label="Seno", color="blue", linewidth=2)
plt.plot(x, coseno, label="Coseno", color="red", linewidth=2)
plt.plot(x, tangente, label="Tangente", color="green", linewidth=2)

# Título del gráfico y etiquetas de los ejes
plt.title("Funciones Trigonométricas: Seno, Coseno y Tangente")
plt.xlabel("Radianes")
plt.ylabel("Valor")

# Limitando el eje Y para la tangente
plt.ylim(-10, 10)

# Añadiendo la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

## Ejercicio 4

Muestra un gráfico de barras comparando dos conjuntos de datos elegidos por tí:

```python
import json

# Cargando los datos del archivo JSON
with open("./datasets/kaggle/EURO_WOMEN_2019.json", "r") as file:
    data = json.load(file)

# Cargando los datos del archivo JSON para el EURO MEN 2020
with open("./datasets/kaggle/EURO_MEN_2020.json", "r") as file:
    data_men = json.load(file)

# Extrayendo información básica para entender la estructura de los datos de los hombres
equipos_casa_men = [partido["home_team"]["home_team_name"] for partido in data_men]
equipos_fuera_men = [partido["away_team"]["away_team_name"] for partido in data_men]
goles_casa_men = [partido["home_score"] for partido in data_men]
goles_fuera_men = [partido["away_score"] for partido in data_men]
goles_casa = [partido["home_score"] for partido in data]
goles_fuera = [partido["away_score"] for partido in data]

# Sumarizando la cantidad total de goles en ambos torneos para comparar
total_goles_women = sum(goles_casa) + sum(goles_fuera)
total_goles_men = sum(goles_casa_men) + sum(goles_fuera_men)

total_goles_women, total_goles_men


# Creando el gráfico de barras para comparar el total de goles entre los torneos de mujeres y hombres
plt.figure(figsize=(8, 5))
bar_labels = ["EURO Women 2019", "EURO Men 2020"]
bar_values = [total_goles_women, total_goles_men]
plt.bar(bar_labels, bar_values, color=["pink", "lightblue"])
plt.title("Comparación de Goles Totales entre EURO Women 2019 y EURO Men 2020")
plt.ylabel("Total de Goles")
plt.show()
```

## Ejercicio 5

Muestra un gráfico de barras comparando tres conjuntos de datos elegidos por tí:

```python
with open("./datasets/kaggle/FCB/FCB2004.json", "r") as file:
    data_2004 = json.load(file)

with open("./datasets/kaggle/FCB/FCB2010.json", "r") as file:
    data_2010 = json.load(file)

with open("./datasets/kaggle/FCB/FCB2015.json", "r") as file:
    data_2015 = json.load(file)

total_goles_casa_2004 = sum(
    [
        partido["home_score"]
        for partido in data_2004
        if partido["home_team"]["home_team_name"] == "Barcelona"
    ]
)
total_goles_casa_2010 = sum(
    [
        partido["home_score"]
        for partido in data_2010
        if partido["home_team"]["home_team_name"] == "Barcelona"
    ]
)
total_goles_casa_2015 = sum(
    [
        partido["home_score"]
        for partido in data_2015
        if partido["home_team"]["home_team_name"] == "Barcelona"
    ]
)

temporadas = ["2004", "2010", "2015"]
goles_casa = [total_goles_casa_2004, total_goles_casa_2010, total_goles_casa_2015]

plt.figure(figsize=(8, 6))
plt.bar(temporadas, goles_casa, color=["gold", "blue", "red"])
plt.title("Goles Marcados en Casa por el FC Barcelona")
plt.xlabel("Temporada")
plt.ylabel("Total de Goles en Casa")
plt.xticks(rotation=45)
plt.show()
```

## Ejercicio 10

Hay una página web llamada [Kaggle](https://www.kaggle.com/datasets) donde se publican distintos datasets y trabajos de machine learning sobre los mismos. Descárgate un dataset de tu elección y muestra datos relevantes del mismo utilizando las gráficas vistas hasta ahora:

```python
import json

with open("./datasets/kaggle/EURO_WOMEN_2019.json", "r") as file:
    data = json.load(file)

total_partidos = len(data)
primer_partido = data[0]
equipos_casa = [partido["home_team"]["home_team_name"] for partido in data]
equipos_fuera = [partido["away_team"]["away_team_name"] for partido in data]
goles_casa = [partido["home_score"] for partido in data]
goles_fuera = [partido["away_score"] for partido in data]

total_partidos, primer_partido.keys(), set(equipos_casa), set(equipos_fuera), sum(
    goles_casa
), sum(goles_fuera)

plt.figure(figsize=(12, 6))
plt.hist(
    goles_casa, bins=range(0, max(goles_casa) + 2), alpha=0.5, label="Goles en Casa"
)
plt.hist(
    goles_fuera,
    bins=range(0, max(goles_fuera) + 2),
    alpha=0.5,
    label="Goles Fuera de Casa",
)
plt.title("Distribución de Goles en Casa vs. Fuera de Casa")
plt.xlabel("Goles")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(goles_casa, goles_fuera, alpha=0.5, color="green")
plt.title("Goles en Casa vs. Fuera de Casa por Partido")
plt.xlabel("Goles en Casa")
plt.ylabel("Goles Fuera de Casa")
plt.grid(True)
plt.show()
```

## Ejercicio 11

Sobre el mismo dataset anterior u otro nuevo, muestra tres gráficas distintas a las aprendidas hasta el momento. Para ello puedes consultar la web de la [documentación de Matplotlib](https://matplotlib.org/stable/gallery/index.html):

```python
plt.figure(figsize=(8, 5))
bar_labels = ["Goles en Casa", "Goles Fuera de Casa"]
bar_values = [sum(goles_casa), sum(goles_fuera)]
plt.bar(bar_labels, bar_values, color=["skyblue", "lightgreen"])
plt.title("Total de Goles en Casa vs. Fuera de Casa")
plt.ylabel("Total de Goles")
plt.show()

plt.figure(figsize=(8, 5))
plt.boxplot([goles_casa, goles_fuera], labels=["Goles en Casa", "Goles Fuera de Casa"])
plt.title("Distribución de Goles en Casa vs. Fuera de Casa")
plt.ylabel("Goles")
plt.show()

goles_casa_acumulados = np.cumsum(goles_casa)
goles_fuera_acumulados = np.cumsum(goles_fuera)

plt.figure(figsize=(10, 6))
plt.fill_between(
    range(len(goles_casa)),
    goles_casa_acumulados,
    color="skyblue",
    alpha=0.4,
    label="Goles en Casa Acumulados",
)
plt.fill_between(
    range(len(goles_fuera)),
    goles_fuera_acumulados,
    color="lightgreen",
    alpha=0.4,
    label="Goles Fuera de Casa Acumulados",
)
plt.title("Acumulación de Goles a lo Largo de los Partidos")
plt.xlabel("Número de Partido")
plt.ylabel("Acumulación de Goles")
plt.legend()
plt.show()
```
