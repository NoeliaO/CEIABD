## Ejercicio 6

Crea un DataFrame a partir de una matriz con valores definidos por tí. Estos valores deberán incluir valores nulos. Luego muestra la cantidad de valores nulos por pantalla y modifica los valores nulos por la media que tenga dicha columna sin incluir los nulos. Muestra el resultado por pantalla:

```python
matriz_con_nulos = np.array(
    [[23, np.nan, 30], [45, 22, np.nan], [np.nan, 18, 25], [32, 20, 27]]
)

df_con_nulos = pd.DataFrame(
    matriz_con_nulos, columns=["Edad Luke", "Edad Leia", "Edad Han"]
)

# Valores nulos por columna
valores_nulos = df_con_nulos.isnull().sum()

# Modificando los valores nulos por la media
df_sin_nulos = df_con_nulos.fillna(df_con_nulos.mean())

valores_nulos, df_sin_nulos
```

## Ejercicio 9

Investiga qué es la correlación entre columnas. Cuando lo hayas entendido, crea una matriz 10x3 con datos manuales que entre la primera y segunda columna de una correlación de 1 en ambos sentidos. Muestra la matriz y su correlación entre columnas por pantalla:

```python
# Datos manuales
valores = np.arange(1, 11)
columna1 = valores
columna2 = valores * 2
# Datos aleatorios
columna3 = np.random.rand(10)

matriz = np.column_stack((columna1, columna2, columna3))

# Conversión a DataFrame
df_manual = pd.DataFrame(matriz, columns=["Columna1", "Columna2", "Columna3"])

# Calculo de la correlación
correlacion = df_manual.corr()

matriz, correlacion
```

### Explicación de la correlación
La correlación entre columnas de un conjunto de datos mide la relación lineal entre las variables representadas por esas columnas. El coeficiente de correlación varía entre -1 y 1, donde:

- **1** indica una correlación positiva perfecta (cuando una variable aumenta, la otra también lo hace en la misma proporción).
- **-1** indica una correlación negativa perfecta (cuando una variable aumenta, la otra disminuye en la misma proporción).
- **0** indica que no hay correlación lineal entre las variables.

Para obtener una correlación de 1 entre dos columnas, los valores en una deben ser una transformación lineal directa de los otros, por ejemplo, multiplicando o sumando un constante (pero sin cambiar el orden relativo de los valores).
