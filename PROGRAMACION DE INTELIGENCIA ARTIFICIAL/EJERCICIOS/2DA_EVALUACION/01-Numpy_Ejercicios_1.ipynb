## Ejercicio 9

Crea una matriz de 3x6 con valores aleatorios. Después muestra por pantalla las siguientes convinaciones:
*   La primera fila entera
*   La primera columna entera
*   Todos los valores entre las columnas 2 y 4
*   Todos los valores de las filas 1 y 2
*   Todos los valores entre las filas 2 y 3 y las columnas 5 y 6

```python
# Creando una matriz de 3x6 con valores aleatorios
matriz_aleatoria_3x6 = np.random.rand(3, 6)

# Mostrando la matriz creada
print("Matriz aleatoria 3x6:")
print(matriz_aleatoria_3x6)

# Extrayendo la primera fila entera
primera_fila = matriz_aleatoria_3x6[0, :]

# Extrayendo la primera columna entera
primera_columna = matriz_aleatoria_3x6[:, 0]

# Extrayendo todos los valores entre las columnas 2 y 4 (recordando que Python indexa desde 0)
columnas_2_a_4 = matriz_aleatoria_3x6[:, 1:4]

# Extrayendo todos los valores de las filas 1 y 2
filas_1_y_2 = matriz_aleatoria_3x6[0:2, :]

# Extrayendo todos los valores entre las filas 2 y 3 y las columnas 5 y 6
filas_2_y_3_columnas_5_y_6 = matriz_aleatoria_3x6[1:3, 4:6]


print(f"Primera fila: {primera_fila},\nPrimera columna: {primera_columna},\nColumnas de la 2 a la 4: {columnas_2_a_4},\Filas 1 y 2: {filas_1_y_2},\nFilas 2 y 3 & Columnas 5 y 6: {filas_2_y_3_columnas_5_y_6}")
```

## Ejercicio 11

Crea dos matrices 3x3, una con valores aleatorios y otra de identidad. Realiza las cuatro operaciones básicas entre ellas (suma, resta, multiplicación y división). Por último, calcula la raíz cuadrada de matriz de identidad.

```python
aleatoria = np.random.rand(3, 3)

mIdentidad = np.eye(3)

suma = aleatoria + mIdentidad
resta = aleatoria - mIdentidad
multiplicacion = np.dot(aleatoria, mIdentidad)
division = aleatoria / mIdentidad


r2identidad = np.sqrt(mIdentidad)

print(f"Suma: {suma},\nResta: {resta},\nMultiplicación: {multiplicacion},\nDivisión: {division},\nRaíz Cuadrada: {r2identidad}")
```

## Ejercicio 12

Sin utilizar las funciones std, mean y median de NumPy, programa dichas funcionalidades:

```python
def calcular_media(array):
    return sum(array) / len(array)

def calcular_mediana(array):
    array_ordenado = sorted(array)
    n = len(array_ordenado)
    mitad = n // 2
    if n % 2 == 0:
        return (array_ordenado[mitad - 1] + array_ordenado[mitad]) / 2
    else:
        return array_ordenado[mitad]

def calcular_desviacion_estandar(array):
    media = calcular_media(array)
    varianza = sum((x - media) ** 2 for x in array) / len(array)
    return varianza ** 0.5

array_ejemplo = np.array([1, 2, 3, 4, 5])

media = calcular_media(array_ejemplo)
mediana = calcular_mediana(array_ejemplo)
desviacion_estandar = calcular_desviacion_estandar(array_ejemplo)

media, mediana, desviacion_estandar
print(f"Media: {media},\nMediana: {mediana},\nDesviación estándar: {desviacion_estandar}")
```
